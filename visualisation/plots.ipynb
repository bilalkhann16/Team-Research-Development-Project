```r
data <- read.csv("//Users/ehzamk/Desktop/archive/epidemic/clusters.csv")
```

```r
head(data)
```

```r
summary(data)
```

```r
x <- data$tests        # Number of tests conducted (independent variable)
y <- data$cases_total  # Total cases reported (dependent variable)

plot(
  x, 
  y, 
  main = "Scatter Plot: Tests Conducted vs Total Cases Reported",
  xlab = "Number of Tests Conducted",                            
  ylab = "Total Cases Reported",                                  
  pch = 19,                                                       
  frame = TRUE                                                    
)
model <- lm(y ~ x, data = data)
abline(model, col = "blue")
```

```r
# 100 rows for clarity
set.seed(123)
sample_data <- data[sample(1:nrow(data), 100), ]

x <- sample_data$tests      
y <- sample_data$cases_total

#X and Y axis to 0-500

plot(
  x, 
  y, 
  main = "Scatter Plot (Zoomed): Tests vs Cases", 
  xlab = "Number of Tests Conducted",            
  ylab = "Total Cases Reported",                 
  pch = 19,                                      
  col = rgb(0, 0, 1, 0.5),                       
  frame = TRUE,                                  
  xlim = c(0, 500),                             
  ylim = c(0, 500)
)

# Add a regression line
model <- lm(y ~ x, data = sample_data)
abline(model, col = "red")            
```
